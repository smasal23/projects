# 📊 ERIA — Impact Analysis (Phase 5 Output)

## Objective
Interpret structured LLM outputs and generate stakeholder-level insights.

## Focus Areas
- Summary interpretation
- Stakeholder impact breakdown
- Risk assessment
- Timeline analysis

```python
import os
import sys
import json

sys.path.append(os.path.abspath(".."))

from src.utils.config import BASE_DIR
from src.utils.helpers import load_json

from src.analysis.summarizer import Summarizer
from src.analysis.stakeholder import StakeholderAnalyzer
from src.analysis.risk_analyzer import RiskAnalyzer
from src.analysis.chronology import ChronologyAnalyzer
```

## 📂 Load LLM Analysis Output

```python
input_path = os.path.join(
    BASE_DIR,
    "artifacts",
    "llm_outputs",
    "eria_analysis.json"
)

data = load_json(input_path)

print(data.keys())
```

## 🧾 Summary

```python
summary = Summarizer.extract(data)

print(summary)
```

## 👥 Stakeholder Impact Analysis

```python
stakeholders = StakeholderAnalyzer.extract(data)

if not stakeholders:
    print("No stakeholder data found")

for role, details in stakeholders.items():
    print(f"\n--- {role.upper()} ---")
    if not details:
        print("No data")
    else:
        for item in details:
            print(f"- {item}")
```

## ⚠️ Risk Analysis

```python
risks = RiskAnalyzer.extract(data)

for r in risks:
    print("\nRisk:", r.get("risk"))
```

## ⏳ Impact Timeline

```python
timeline = ChronologyAnalyzer.extract(data)

print("Short Term:", timeline.get("short_term"))
print("\nMedium Term:", timeline.get("medium_term"))
print("\nLong Term:", timeline.get("long_term"))
```

## 📌 Key Insights

### Strengths
- Clear summary simplifies regulation complexity
- Good stakeholder segmentation
- Risks identified with severity levels

### Limitations
- Timeline may lack factual grounding
- Some outputs may be generic
